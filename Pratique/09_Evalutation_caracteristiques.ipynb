# Evalutation des caractéristiques
Quelques conseils:
* Toujours utiliser des fonctions non linéaires avec un plafond. Dans le cas contraire plusieurs problèmes:
  * Les valeurs aberrantes peuvent fausser assez gravement les résultats
  * Un critère risque d'être surdominant dans certains contextes
* Si une fonction linéaire fait vraiemnt sens, fixer malgré tou un plafond
* Attention aux divisions par 0
* Il faut gérer les valeurs manquantes
* Il faut gérer les valeurs aberrantes: date de publication dans le futur par exemple

```python
# Bibliothèques utilisées
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Plus la valeur est grande, moins le résultat de l'évaluation est élevé
Ce type d'évaluation est par exemple utile pour évaluer la date de publication. Une racine carrée, une puissance ou un logarithme naturel permettent souvent d'adapter les données

$$
\frac{max\_value}{année\_actuelle - année\_publication + 1} = évaluation
$$

```python
x = np.arange(15)
y_data = []
g_title = []

y_data.append([1000/(e+1) for e in x])
g_title.append('Max_value/(value+1)')

y_data.append([1000/np.sqrt(e+1) for e in x])
g_title.append('Max_value/sqrt(value+1)')

y_data.append([1000/(e+1)**2 for e in x])
g_title.append('Max_value/(value+1)**2')
```

```python
fig, axis = plt.subplots(1, 3, figsize=(18,5))

for i, y in enumerate(y_data):
    sns.lineplot(x=x, y=y, ax=axis[i])
    axis[i].set_title(g_title[i])
    
```

### Plus la valeur est grande, plus le résultat de l'évaluation est élevé
Ce type d'évaluation est par exemple utile pour évaluer le nombre de prêts. Une racine carrée, une puissance ou un logarithme naturel permettent souvent d'adapter les données

$$
\frac{nb\_prêts}{nb\_prêts+1} max\_value = évaluation
$$

```python
x = np.arange(15)
y_data = []
g_title = []

y_data.append([e / (e + 1) * 1000 for e in x])
g_title.append('value / (value+1) x Max_value')

y_data.append([np.sqrt(e / (e + 1)) * 1000 for e in x])
g_title.append('sqrt(value / (value+1)) x Max_value')

y_data.append([(e / (e + 1))**2 * 1000 for e in x])
g_title.append('(value / (value+1))**2 x Max_value')
```

```python
fig, axis = plt.subplots(1, 3, figsize=(18,5))

for i, y in enumerate(y_data):
    sns.lineplot(x=x, y=y, ax=axis[i])
    axis[i].set_title(g_title[i])
```
